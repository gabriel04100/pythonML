```python
from sklearn import datasets
iris = datasets.load_iris()

print(iris['data'].max(axis=0))
print(iris['data'].min(axis=0))
```

```python
import random
#random.uniform(iris['data'].min(axis=0)[0],iris['data'].max(axis=0)[0])
lign,col=iris['data'].shape
```

```python
def random_centroids(data):
    minimums=data.min(axis=0)
    maximums=data.max(axis=0)
    centroids=[]
    for i in range(len(minimums)):
        centroids.append(random.uniform(minimums[i],maximums[i]))
        
    return centroids
```

```python
random_centroids(iris['data'])
```
