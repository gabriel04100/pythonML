```python
import random
import numpy as np
```

```python
type(np.array([]))


def bootstrap(data:numpy.ndarray,number_sample:int=1)->numpy.ndarray :
    result=np.zeros((number_sample,len(data)))
    for i in range(number_sample):
        for j in range(len(data)):
            result[i,j]=random.choice(data)
    return result



def k_fold(data:numpy.ndarray,k_splits:int) ->numpy.ndarray:
    size= int(len(data)/k_splits)
    result=np.zeros((k_splits,int(size)))

    return np.array([data[x:x+size] for x in range(0, len(data), size)])

    
    
```

```python
data=np.array([1,2,3,5,6,7])
bootstrap(data,5)
k_fold(data,3)
```
